```python
from opencc import OpenCC
```

```python
t2s = OpenCC('t2s')
s2t = OpenCC('s2t')
```

```python
tw = '台灣萬歲'

cha = t2s.convert(tw)

print(cha)
```

```python
tw2 = s2t.convert(cha)

print(tw2)
```
